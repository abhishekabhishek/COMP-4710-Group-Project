## To do : Extract the mbid of the the previous 10000 tracks listened to by the user and all the level 1 friends which have listened to more than 10000 tracks.

```python
import json
import requests

def get_track_array(user_name):
    
    threshold = 10000
    
    # Determine whether the user actually has listened to more than 10000 tracks
    tracks_initial_request = 'http://ws.audioscrobbler.com/2.0/?method=user.getrecenttracks&user=' + user_name + '&api_key=' + API_KEY + '&format=json' + '&limit=1'
    r = requests.get(tracks_initial_request)
    print('Printing the response for the first request for user : ' + user_name)
    print(r.text)
    
    # Use json.loads to initialize the dictionary
    request_output = json.loads(r.text)
    
    # Get the number of tracks the user has listened to
    num_tracks = int(request_output['recenttracks']['@attr']['total'])
    
    # If the number of tracks is less than threshold
    if num_tracks < threshold :
        print('The user has listened to only ' + str(num_tracks) + ' which is lower than ' + str(threshold))
        return None
    else:
        track_array = []
        page = 1
        
        while( len(track_array) < threshold ):
            
            tracks_request = 'http://ws.audioscrobbler.com/2.0/?method=user.getrecenttracks&user=' + user_name + '&api_key=' + API_KEY + '&format=json' + '&limit=200&page=' + str(page)
            r = requests.get(tracks_request)
            
            # Use json.loads to initialize the dictionary
            request_output = json.loads(r.text)
            
            # Extract the information from the dictionary
            for i in range(len(request_output['recenttracks']['track'])):
                track_name = request_output['recenttracks']['track'][i]['name'].replace(" ", "+")
                artist_name = request_output['recenttracks']['track'][i]['artist']['#text'].replace(" ", "+")
                track_array.append((track_name, artist_name))
            
            page += 1
            
        return track_array
```

```python
user_1_track_array = get_track_array(user_1.name)
display(user_1_track_array)
```

```python
user_1_friends = user_1.get_friends()
user_1_friends_track_array = []

# Extract the track array for all the friends and append to the main array
for friend in user_1_friends:
    user_1_friends_track_array.append(get_track_array(friend.name))
```

```python
from json import JSONDecodeError

def map_to_genre(track_array, threshold=50):
    
    mapped_array = []
    unique_tags = []
    track_index = 0
    
    # Map all the tracks in the given array to their respective arrays
    for track_info in track_array:
        
        add = True
        
        track = track_info[0].replace("#", "")
        artist = track_info[1].replace("#", "")
        
        track = track.replace("&", "and")
        artist = artist.replace("&", "and")
        
        if "Featuring" in track:
            track = track[:track.find("Featuring")]
            
        
        try:
            toptag_request = 'http://ws.audioscrobbler.com/2.0/?method=track.gettoptags&track=' + track + '&artist=' + artist + '&api_key=' + API_KEY + '&format=json'
            r = requests.get(toptag_request)
        
            # Use json.loads to initialize the dictionary
            request_output = json.loads(r.text)
            
        except JSONDecodeError:
            print('Unsuccessful request : ' + toptag_request)
            add = False

        if( add ):
            
            try:

                # Extract the current track genre from the dict
                if( len(request_output['toptags']['tag']) > 0 ):

                    if(len(unique_tags) < threshold):

                        tag = request_output['toptags']['tag'][0]['name']

                        mapped_array.append(tag)

                        if tag not in unique_tags: 
                            unique_tags.append(tag)

                    else:

                        if( len(request_output['toptags']['tag']) == 1 ):
                            tag = request_output['toptags']['tag'][0]['name']
                            mapped_array.append(tag)
                            if tag not in unique_tags: 
                                unique_tags.append(tag)

                        else:

                            index = 0
                            tag_added = False

                            while(index < len(request_output['toptags']['tag']) and not tag_added):

                                tag = request_output['toptags']['tag'][index]['name']

                                if tag in unique_tags:
                                    mapped_array.append(tag)
                                    tag_added = True

                                index += 1

                            if not tag_added:
                                
                                tag = request_output['toptags']['tag'][0]['name']
                                mapped_array.append(tag)
                                
                                if tag not in unique_tags:
                                    unique_tags.append(tag)
                                
                                tag_added = True

            except KeyError:
                print("Key not found for track = " + track + " and artist = " + artist)
                print(request_output)
        
        track_index += 1

    return mapped_array, unique_tags
```
